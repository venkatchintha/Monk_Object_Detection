```
from train_detector import Detector
```

```
gtf = Detector();
```

```


root_dir = "/content/drive/My Drive/Kaggle/";
coco_dir = "nexet";
set_dir = "nexet_2017_1";
```

```
gtf.Train_Dataset(root_dir, coco_dir, set_dir, batch_size=4,image_size=300, num_workers=3)
```
